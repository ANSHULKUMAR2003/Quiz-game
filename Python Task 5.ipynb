```python
import random

# Quiz questions with options and correct answers
questions = [
    {
        "question": "What is the capital of France?",
        "options": ["A) London", "B) Paris", "C) Rome", "D) Berlin"],
        "correct_answer": "B"
    },
    {
        "question": "Which planet is closest to the Sun?",
        "options": ["A) Earth", "B) Mars", "C) Mercury", "D) Venus"],
        "correct_answer": "C"
    },
    {
        "question": "Who painted the Mona Lisa?",
        "options": ["A) Leonardo da Vinci", "B) Pablo Picasso", "C) Vincent van Gogh", "D) Michelangelo"],
        "correct_answer": "A"
    },
    # Add more questions here
]

def shuffle_options(options):
    # Randomly shuffle the options
    random.shuffle(options)
    return options

def display_question(question_data):
    print(question_data["question"])
    for option in shuffle_options(question_data["options"]):
        print(option)

def get_player_answer():
    while True:
        answer = input("Enter the letter corresponding to your answer: ").upper()
        if answer in ["A", "B", "C", "D"]:
            return answer
        else:
            print("Invalid input. Please enter A, B, C, or D.")

def evaluate_answer(player_answer, correct_answer):
    if player_answer == correct_answer:
        print("Correct!\n")
        return True
    else:
        print("Incorrect!\n")
        return False

def quiz_game(questions):
    score = 0
    total_questions = len(questions)

    print("Welcome to the Quiz Game!")
    print("You will be presented with multiple-choice questions. Enter the letter corresponding to your answer.")

    for idx, question in enumerate(questions, start=1):
        print(f"\nQuestion {idx}/{total_questions}:")
        display_question(question)
        player_answer = get_player_answer()
        if evaluate_answer(player_answer, question["correct_answer"]):
            score += 1

    print(f"\nQuiz completed! Your score: {score}/{total_questions}")

if __name__ == "__main__":
    quiz_game(questions)
```
